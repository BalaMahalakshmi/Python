```python
import matplotlib
print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,8])
ypoints = np.array([0,200])
plt.plot(xpoints, ypoints)
plt.show()
 
```

```python
xpoints = np.array([1,8])
ypoints = np.array([3,10])
plt.plot(xpoints, ypoints)
plt.show()
```

without line

```python
xpoints = np.array([1,8])
ypoints = np.array([3,10])
plt.plot(xpoints, ypoints, 'o') #o means rings
plt.show()
```

multiple points

```python
x  = np.array([1,2,6,8,1])
y = np.array([3,8,1,8,3])
plt.plot(x, y)
plt.show()
```

default x points

```python
y = np.array([10,2,4,5,6,10,3,4,5,6,1])
plt.plot(y)
plt.show()
```

```python
y = np.array([10,2,3.5,5,6.5])
plt.plot(y, marker='*') # use 'o,.,x,X,+,P,s,D,d(thin),H,h,V(triangle down),^(triangle up),<,>,1,2,3,4,|,-'
plt.show()
```

format strings 

```python
y = np.array([2,4.5,6,8.5,10.5])
plt.plot(y, 's:b') #marker|line|color, 'm:c'
plt.show()
```

color reference: r,g,c,y,k,w,b,m
line reference: -, :, --, -.

```python
y = np.array([10,8,6,4,2,2,4,6,8,10])
plt.plot(y, 'H-.c') 
plt.show()
```

```python
y = np.array([3,8,1,5])
plt.plot(y,'o', ms=16)
plt.show()
```

```python
y = np.array([6,4,8,3])
plt.plot(y,'s:', mfc='k')
plt.show()
```

```python
y = np.array([6,4,8,3])
plt.plot(y,'s:', mec='y')
plt.show()
```

```python
y = np.array([6,5,2,3,8])
plt.plot(y,'s:', mfc='k', ms=8, mec='b')
plt.show()
```

```python
y = np.array([6,4,8,3])
plt.plot(y,'s--', mfc='hotpink',mec='hotpink',ms=12)
plt.show()
```

```python
y = np.array([2,6,8,4,2,5])
plt.plot(y,'h:', color='m', linewidth='8.4')
plt.show()
```

```python
x = np.array([2,5,7,4,1])
y = np.array([2,4,6,4,6,2])
plt.plot(x)
plt.plot(y)
plt.show()
```
