```python
import numpy as np
import matplotlib.pyplot as plt

# Create time points
x = np.linspace(0, 2*np.pi, 500)

# Create synthetic heartbeat-like signal
y = np.sin(3*x) * np.exp(-0.3*x) + np.sin(15*x) * 0.2

# Make the spikes sharper to look more like ECG
y[100:105] += 1.5
y[300:305] += 1.5

plt.figure(figsize=(10, 4))
plt.plot(x, y, color='red')
plt.title("Heartbeat of both in Heart")
plt.xlabel("ABI")
plt.ylabel("MAHA")
plt.grid(True)
plt.show()
```
